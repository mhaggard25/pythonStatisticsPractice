In this project, I will analyze data from qualified hitters from the 2024 MLB season and determine the mean, median, mode, range, quartiles, IQR, and outliers for Batting Average, On-Base %, and Slugging %. I will also graph the data using a box and whisker plot and label the outliers if any exist.

This project will help me better see who performed excepltionally well and exceptionally bad throughout this season in regard to these three statistics. It will also help me practice using Python and its libraries to read and analyze data efficiently and quickly. 

## Step 1.
The first step in this project will be to open the .csv file and look at the layout of the data. This is important because it will help me understand how to read in the data properly. I will be able to see immediately what parts of the data that I want and what parts are not important to this exercise.

After opening the .csv file, I can see that all three stats that I need are already calculated. Their labels are 'ba', 'slg', and 'obp'. 

Now, I can setup my python code to read in the data.

```python
# import the libraries that we need.
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np

# create a variable for the filename
str_filename = '2024_QualifiedHitters.csv'

file_header = open(str_filename)
csv_reader = csv.reader(file_header)

csv_header = next(csv_reader)
csv_header
```

## Step 2.

Next, I will go ahead and close the file after i've read the header information with the csv library. Then, I will read in the data from the file using the pandas library. (Reading in the header this way seems a little unecessary with this particular file, but if the file had several lines at the top of the table, it would be easier to get the header lines that I need.)

Next, I will read the data from the table using the pandas library, ensuring that I skip the first line in this particular file since I created a list of headers in the previous step.

```python
# close the file 
file_header.close()

# read the data from the file using the pandas library.
df_sig = pd.read_csv(str_filename, header=None, skiprows=1, names=csv_header)
df_sig
```

## Step 3.

Determine the mean, median, mode, and range of the 3 stats that I am concerned with: ba, slg, and obp.

Pandas has methods that will determine these values without having to save data to external lists or variables.

```python
# Pandas has built in functions for mean, median, and mode so using the statistics library isn't necessary here.

# Print the calculations on BA
print('****************')
print('Batting Average')
print('****************')

print('The mean of the Batting Averages in 2024 was:', round(df_sig['ba'].mean(), 3))
print('The median of the Batting Averages in 2024 was:', round(df_sig['ba'].median(), 3))
print('The mode of the Batting Averages in 2024 was:\n' + str(round(df_sig['ba'].mode(), 3)))
print('The range of the Batting Averages in 2024 was:', round(df_sig['ba'].max(), 3) - round(df_sig['ba'].min(), 3))

# Print the calculations on OBP
print('\n****************')
print('On-Base %')
print('****************')

print('\nThe mean of the On-Base Percentages in 2024 was:', round(df_sig['obp'].mean(), 3))
print('The median of the On-Base Percentages in 2024 was:', round(df_sig['obp'].median(), 3))
print('The modes of the On-Base Percentages in 2024 was:\n' + str(round(df_sig['obp'].mode(), 3)))
print('The range of the On-Base Percentages in 2024 was:', round(df_sig['obp'].max(), 3) - round(df_sig['obp'].min(), 3))

# Pring the calculations on Slugging %
print('\n****************')
print('Slugging Percentage')
print('****************')

print('\nThe mean of the Slugging Percentages in 2024 was:', round(df_sig['slg'].mean(), 3))
print('The median of the Slugging Percentages in 2024 was:', round(df_sig['slg'].median(), 3))
print('The modes of the Slugging Percentages in 2024 was:\n' + str(round(df_sig['slg'].mode(), 3)))
print('The range of the Slugging Percentages in 2024 was:', round(df_sig['slg'].max(), 3) - round(df_sig['slg'].min(), 3))
```
